# 01 EDA
Load PlantVillage class folders, plot class distribution, inspect samples, summarize image sizes, and flag unreadable files.

```python
from pathlib import Path
from collections import Counter
import cv2
import matplotlib.pyplot as plt

raw_dir = Path('../data/raw/PlantVillage')
image_paths = [p for p in raw_dir.rglob('*') if p.suffix.lower() in {'.jpg', '.jpeg', '.png', '.bmp', '.webp'}]
labels = [p.parent.name for p in image_paths]
counts = Counter(labels)
print(f'Total images: {len(image_paths)}')
print(f'Total classes: {len(counts)}')
for name, count in counts.most_common():
    print(name, count, '<50 images' if count < 50 else '')
```

```python
plt.figure(figsize=(14, 6))
plt.bar(counts.keys(), counts.values())
plt.xticks(rotation=90, fontsize=8)
plt.title('Class distribution')
plt.tight_layout()
plt.show()
```

```python
sizes = []
bad = []
for path in image_paths:
    img = cv2.imread(str(path))
    if img is None:
        bad.append(path)
        continue
    h, w = img.shape[:2]
    sizes.append((w, h, w / h))
print(f'Unreadable files: {len(bad)}')
print(bad[:10])
if sizes:
    widths, heights, ratios = zip(*sizes)
    print('Width min/mean/max:', min(widths), sum(widths)/len(widths), max(widths))
    print('Height min/mean/max:', min(heights), sum(heights)/len(heights), max(heights))
    print('Aspect ratio min/mean/max:', min(ratios), sum(ratios)/len(ratios), max(ratios))
```

```python
classes = list(counts)[:9]
fig, axes = plt.subplots(3, 3, figsize=(9, 9))
for ax, class_name in zip(axes.ravel(), classes):
    sample = next((raw_dir / class_name).glob('*'))
    img = cv2.cvtColor(cv2.imread(str(sample)), cv2.COLOR_BGR2RGB)
    ax.imshow(img)
    ax.set_title(class_name, fontsize=8)
    ax.axis('off')
plt.tight_layout()
plt.show()
```
